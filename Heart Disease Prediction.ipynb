```python
x=data_set.iloc[:,1:13]
y=data_set.iloc[:,13]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lrmodel=LogisticRegression(random_state=0)
lrmodel.fit(x_train,y_train)
y_predlr=lrmodel.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm_lr=confusion_matrix(y_test,y_predlr)
accuracy_score_lr=accuracy_score(y_test,y_predlr)
```

```python
plt.figure(figsize=(15,7))
plt.title('LogisticRegression Confusion Matrix',)
sns.heatmap(cm_lr,annot=True,cmap='Blues',fmt='d',cbar=False)
plt.show()
```

```python
print('Accuracy of Logistic Regression:{}'.format(accuracy_score_lr))
```

```python
#KNN
```

```python
from sklearn.neighbors import KNeighborsClassifier
knnmodel=KNeighborsClassifier()
knnmodel.fit(x_train,y_train)
y_predknn=knnmodel.predict(x_test)
```

```python
cm_knn=confusion_matrix(y_test,y_predknn)
accuracy_score_knn=accuracy_score(y_test,y_predknn)
```

```python
plt.figure(figsize=(15,7))
plt.title('K Nearest Neighbours Confusion Matrix')
sns.heatmap(cm_knn,annot=True,cmap='Blues',fmt='d',cbar=False)
plt.show()
```

```python
print('Accuracy of K Neighbors Classifier:{}'.format(accuracy_score_knn))
```

```python
#SVM
```

```python
from sklearn.svm import SVC
svm=SVC(kernel='linear',random_state=0)
svm.fit(x_train,y_train)
y_predsvm=svm.predict(x_test)
```

```python
cm_svm=confusion_matrix(y_test,y_predsvm)
accuracy_score_svm=accuracy_score(y_test,y_predsvm)
```

```python
plt.figure(figsize=(15,7))
plt.title('SVM Confusion Matrix')
sns.heatmap(cm_svm,annot=True,cmap='Blues',fmt='d',cbar=False)
plt.show()
```

```python
print('Accuracy of SVM:{}'.format(accuracy_score_svm))
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
y_prednb=nb.predict(x_test)
```

```python
cm_nb=confusion_matrix(y_test,y_prednb)
accuracy_score_nb=accuracy_score(y_test,y_prednb)
```

```python
plt.figure(figsize=(15,7))
plt.title('Naive Bayes Confusion Matrix')
sns.heatmap(cm_nb,annot=True,cmap='Blues',fmt='d',cbar=False)
plt.show()
```

```python
print('Accuracy of Naive Bayes:{}'.format(accuracy_score_nb))
```

```python
#Decision Tree
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier(criterion='gini',random_state=0)
tree.fit(x_train,y_train)
y_predtree=tree.predict(x_test)
```

```python
cm_tree=confusion_matrix(y_test,y_predtree)
accuracy_score_tree=accuracy_score(y_test,y_predtree)
```

```python
plt.figure(figsize=(15,7))
plt.title('Decision Tree Confusion Matrix')
sns.heatmap(cm_tree,annot=True,cmap='Blues',fmt='d',cbar=False)
plt.show()
```

```python
print('Accuracy of Decision Trees:{}'.format(accuracy_score_tree))
```

```python
#Random Forest 
```

```python
from sklearn.ensemble import RandomForestClassifier
forest=RandomForestClassifier(criterion='gini',n_estimators=10)
forest.fit(x_train,y_train)
y_predforest=forest.predict(x_test)
```

```python
cm_forest=confusion_matrix(y_test,y_predforest)
accuracy_score_forest=accuracy_score(y_test,y_predforest)
```

```python
plt.figure(figsize=(15,7))
plt.title('Random Forest Confusion Matrix')
sns.heatmap(cm_forest,annot=True,cmap='Blues',fmt='d',cbar=False)
plt.show()
```

```python
print('Accuracy of Random Forest:{}'.format(accuracy_score_forest))
```

```python
classification_models=['Linear Regression','K Nearest Neighbours','SVM','Naive Bayes','Decision Tree','Random Forest']
model_accuracy=[accuracy_score_lr,accuracy_score_knn,accuracy_score_svm,accuracy_score_nb,accuracy_score_tree,accuracy_score_forest]
```
